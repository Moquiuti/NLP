```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def nuvem_palavras(df, coluna_texto, sentimento):
    # Filtra o DataFrame pelo sentimento desejado
    texto_sentimento = df[df['sentimento'] == sentimento][coluna_texto]

    # Junta todos os textos em uma única string
    texto_unico = ' '.join(texto_sentimento.astype(str))

    # Cria a nuvem de palavras
    nuvem = WordCloud(
        width=800,
        height=400,
        background_color='white',
        collocations=False
    ).generate(texto_unico)

    # Exibe a nuvem
    plt.figure(figsize=(12, 6))
    plt.imshow(nuvem, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Nuvem de palavras - Sentimento {sentimento}')
    plt.show()
```

```python

nuvem_palavras(df, 'avaliacao', 'negativo')
```
